    We need to sort it, for that we will make a function.

```
def sort(obj):
    M = []
    for j in ast.literal_eval(obj):
        M.append(j['name'])
    return M
```

ast.literal_eval is a function to change string of list to list.

```
movies_set['genres'].apply(sort)
```

```
movies_set['keywords'].apply(sort)
```

```
movies_set['genres'] = movies_set['genres'].apply(sort)
movies_set['keywords'] = movies_set['keywords'].apply(sort)
```

    In the 'cast' column, we just need first three of their actual names without their character's name.

    For that we will make another function, in which we will sort the data on the basis of name.

```
def sort_name(obj):
    
    M = []
    count = 0
    
    for j in ast.literal_eval(obj):
        if count != 3:
            M.append(j['name'])
            count += 1
        else: 
            break
    return M
```

```
movies_set['cast'].apply(sort_name)
```

```
movies_set['cast'] = movies_set['cast'].apply(sort_name)
```

    In the 'crew' column, we just need director's name.

    To get that we will make a function in which we will sort the data on the basis of roles.

```
def sort_role(obj):
    M = []
    for j in ast.literal_eval(obj):
        if j['job'] == 'Director':
            M.append(j['name'])
            break
    return M
```

```
movies_set['crew'].apply(sort_role)
```

```
movies_set['crew'] = movies_set['crew'].apply(sort_role)
```

**Now we will remove the spacing in between names and surnames to avoid confusion. As their might be some names whose first name is same but second name is different, due to this while executing the code, it might get confusing.**

```
movies_set['genres'] = movies_set['genres'].apply(lambda x:[i.replace(" ","") for i in x])

movies_set['keywords'] = movies_set['keywords'].apply(lambda x:[i.replace(" ","") for i in x])

movies_set['cast'] = movies_set['cast'].apply(lambda x:[i.replace(" ","") for i in x])

movies_set['crew'] = movies_set['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```
new_set['tags'] = new_set['tags'].apply(lambda x:x.lower())
```

*Converted to lower case*

Now we will convert tags in the form of vector. We will use 'Bags of words' technique.

We will use a library scikit learn which is downloaded from net. In that we will use a function CountVectorrizer.

```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

    So in 'cv', we will store top 5000 words which are commanly used in this data set.

    And we will remove stop words. 

   *stop words are words that doesn't contribute in sentence formation, like on, is ,for ,and etc..*

```
cv.fit_transform(new_set['tags']).toarray()
```

*added those words in an array using numpy*

```
vectors = cv.fit_transform(new_set['tags']).toarray()
```

```
len(cv.get_feature_names_out())
```

```
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

*Here we have used steaming to remove words which are almost commmon and make them a single word.*

*Ex: Actor, Acctors*

```
new_set['tags'].apply(stem)
```

```
new_set['tags'] = new_set['tags'].apply(stem)
```

```
cv.get_feature_names()
```

**Now we will calculate the similarity between two values of vector using the comaparision of their cosine values, those movies which have lower cosine value, is likely to be similar and vive-versa.**

```
from sklearn.metrics.pairwise import cosine_similarity
```

```
cosine_similarity(vectors)
```

```
cosine_similarity(vectors).shape
```

    Now we will calculate 5 closest movies to a movie by finding 5 closest cosine value of that movie in vector.

```
similarity = cosine_similarity(vectors)
```

```
def recommend(movie):
    movie_index = new_set[new_set['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print (new_set.iloc[i[0]].title)
```

```
recommend('Batman')
```
